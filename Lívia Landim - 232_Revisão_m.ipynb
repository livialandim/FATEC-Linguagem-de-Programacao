###2. Faça um programa para calcular e apresentar a média de um aluno de ADS, referente a duas provas que foram realizadas. Mostre uma frase se ele foi aprovado ou reprovado.

```
p1 = float(input('Informe a nota obtida na P1: '))
p2 = float(input('Informe a nota obtida na P2: '))

media = (p1 + p2) / 2

if media < 6:
    print(f'Você foi reprovado com uma média {media}.')
else:
    print(f'Você foi aprovado com uma média {media}.')
```

### 3. Faça um programa que calcule e apresente a tabuada do número 7.

### 4. Faça um programa que armazene em um vetor 8 números. Calcule e mostre quantos elementos são positivos.

```
i = 0
numero = []
positivos = 0

for i in range(8) :
    numero.append(float(input('Informe um número: ')))

    if numero[i] > 0:
        positivos = positivos + 1

print(f'A quantidade de números positivos é {positivos}')
```
