Com os dados lidos, armazene na estrutura Futebol (posicao_jogo, altura, peso, imc), calcule o IMC (Índice de Massa Corporal), armazene na estrutura e também em outro arquivo, mas agora chamado ---> futebol_imc**.CSV** <---.
##Fórmula imc = peso / (altura ** 2)

##Baseado no exercício 4 da lista de exercícios de arquivo, gere o gráfico, para isso, use a coluna **imc** e adote a cor orange da biblioteca matplotlib.pyplot as plt
##plt.hist()

```
# Importando as bibliotecas
import matplotlib.pyplot as plt
import pandas as pd

# Definindo a classe jogador
class jogador :
    posicao_jogo = ''
    altura = 0.0
    peso = 0.0
    imc = 0.0

def trabalharArquivos() :
    # Leitura do arquivo 'futebol.txt'
    arquivo = open('futebol.txt', 'r')
    # Criando o arquivo 'futebol_imc.csv'
    arquivoIMC = open('futebol_imc.csv', 'w')

    arquivoIMC.write('Posicao,Altura,Peso,IMC\n')

    for linha in arquivo.readlines() :
        posicao, altura, peso = linha.strip().split(',')

        j = jogador
        j.posicao_jogo = posicao
        j.altura = float(altura)
        j.peso = float(peso)
        # Cálculo do IMC
        j.imc = j.peso / (j.altura ** 2)

        arquivoIMC.write(f'{j.posicao_jogo},{j.altura},{j.peso},{j.imc:.2f}\n')

    # Fechando os arquivos
    arquivo.close()
    arquivoIMC.close()

def main() :
    trabalharArquivos()

    df = pd.read_csv('futebol_imc.csv')

    # Criando o gráfico
    plt.hist(df['IMC'], 5, rwidth = 0.8, color = 'orange')
    plt.title('Índice de Massa Corporal dos jogadores')
    plt.xlabel('IMC')
    plt.ylabel('Quantidade')

    plt.show()

main()
```
