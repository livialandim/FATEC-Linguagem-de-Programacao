```
# Estudando leitura de arquivos
import pandas as pd

# DataFrame é  como uma planilha
df = pd.read_excel('Vendas.xlsx')
print(df['ID Loja'].value_counts()) # Lê a quantidade de repetições de algo, no
# caso, interpretando a tabela, quantidade de produtos

print('-'*50)

print(df) # Apresenta uma tabela de acordo com o documento fornecido

print('-'*50)

print(df.info()) # Mostra as informações do documento

print(df.columns.ravel()) # Mostra o cabeçalho do excel

print('-'*50)

faturamento = df['Valor Final'].sum() # Soma todos os valores da coluna selecionada
print(f'Faturamento: R${faturamento:,.2f}') # A vírgula serve para dividir o número em mil, milhão, bilhão, etc
# Resultado: R$ 2,456,789.00
```

### 6. Crie um programa que carregue o arquivo Propaganda.csv, por meio da biblioteca Pandas. Gere um histograma, referente a coluna TV, utilizando a biblioteca Matplotlib, apresente na cor blueviolet.


```
import matplotlib.pyplot as plt

df = pd.read_csv('Propaganda.csv')

plt.hist(df['TV'], 5, rwidth = 0.8, color = 'blueviolet')
# Opcionais:
plt.title('Propaganda via Televisão')
plt.xlabel('Custo')
plt.ylabel('Quantidade')

plt.show()
```
