### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar

```
class caracProduto :
    codigo = 0
    nome = ''
    preco = 0.0

def cadastrar(vetorProdutos) : # Boa prática : nomear funções com nome de verbo
    for i in range(3) :
        p = caracProduto()

        p.codigo = int(input('Informe o código do produto: '))
        p.nome = input('Informe o nome do produto: ')
        p.preco = float(input('Informe o preço do produto: '))
        vetorProdutos.append(p)

    print()
    print('Cadastro realizado com sucesso!')

    return vetorProdutos

def reajustar(vetorProdutos) :
    for i in range(len(vetorProdutos)) :
        vetorProdutos[i].preco = vetorProdutos[i].preco * 1.1
    print()
    print('Reajuste realizado com sucesso!')

def visualizar(vetorProdutos) :
    print('Código\t\tNome\t\tPreço R$')

    for i in range(len(vetorProdutos)) :
        print(f'{vetorProdutos[i].codigo} \t\t {vetorProdutos[i].nome} \t\t {vetorProdutos[i].preco:.2f}')

def consultar(vetorProdutos) :
    if len(vetorProdutos) > 0:
        encontrado = False
        produtoConsultar = input('Qual o produto a ser encontrado? ')
        for i in range(len(vetorProdutos)) :
            if produtoConsultar == vetorProdutos[i].nome : # Especificar qual atributo
                encontrado = True
                print('Produto encontrado.')
                print(f'Código: {vetorProdutos[i].codigo}; Nome: {vetorProdutos[i].nome}; Preço: R${vetorProdutos[i].preco:.2f}')
                break # Interromper programação
        if encontrado == False : # if not encontrado
            print('O produto não foi encontrado.')
        # Ao avaliar uma variável/função BOOL, é uma boa prática escrever:
        # if encontrado : # O sistema vê como verdadeiro
        # if not encontrado: # É falso
    else :
        print('Não há produtos cadastrados.')

def alterar(vetorProdutos) :
    if len(vetorProdutos) > 0:
        codigoPesquisar = int(input('Qual código deseja encontrar para alteração? '))
        for i in range(len(vetorProdutos)) :
            if codigoPesquisar == vetorProdutos[i].codigo :
                vetorProdutos[i].nome = input('Qual o novo nome? ')
                vetorProdutos[i].preco = float(input('Qual o novo preço? '))
        print('Produto alterado com sucesso.')
    else :
        print('Não há produtos cadastrados.')
    return vetorProdutos

def excluir(vetorProdutos) :
    if len(vetorProdutos) > 0:
        for i in range(len(vetorProdutos)) :
            if i == 0 :
                menorPreco = vetorProdutos[i].preco
                indiceMenor = i
            if menorPreco > vetorProdutos[i].preco :
                menorPreco = vetorProdutos[i].preco
                indiceMenor = i
        vetorProdutos.pop(indiceMenor)
        print('Produto excluido com sucesso.')
    else :
        print('Não há produtos cadastrados.')
    return vetorProdutos
def main() :
    vetorProdutos = []

    print('Menu do sistema')
    print('1 - Cadastro de produtos')
    print('2 - Reajuste de produtos')
    print('3 - Visualização de produtos')
    print('4 - Consultar produtos')
    print('5 - Alterar produtos')
    print('6 - Excluir produto')
    print('7 - Sair')

    escolha = int(input('Qual opção você deseja? '))

    while escolha < 8 :
        if escolha == 1 :
            cadastrar(vetorProdutos)
        elif escolha == 2 :
            reajustar(vetorProdutos)
        elif escolha == 3 :
            visualizar(vetorProdutos)
        elif escolha == 4 :
            consultar(vetorProdutos)
        elif escolha == 5 :
            alterar(vetorProdutos)
        elif escolha == 6 :
            excluir(vetorProdutos)
        elif escolha == 7 :
            print('Fim da execução do sistema')
            break # Interrompe a estrutura de repetição

        print('Menu do sistema')
        print('1 - Cadastro de produtos')
        print('2 - Reajuste de produtos')
        print('3 - Visualização de produtos')
        print('4 - Consultar produtos')
        print('5 - Alterar produtos')
        print('6 - Excluir produto')
        print('7 - Sair')

        escolha = int(input('Qual opção você deseja? '))

main()
```

Menu de opções:

Observação:

```
# Percorra todo os vetor e quando a condição de enontrar um menor saldo for executada, guarde o índice
# Fora do laço de repetição, exclua pelo índice, por exemplo com o comando pop(índice)

class contaCliente :
    numeroConta = 0
    nomeTitular = ''
    saldo = 0.0

def cadastrar(vetorConta) :
    for quantidadeContas in range(15) :
        print('Cadastro da conta ',quantidadeContas,'/15')
        conta = contaCliente()
        conta.numeroConta = int(input('Informe o número da conta: '))
        conta.nomeTitular = input('Informe o nome do titular: ')
        conta.saldo = float(input('Informe o saldo atual da conta: '))
        vetorConta.append(conta)
    return vetorConta

def visualizar(vetorConta) :
    if len(vetorConta) > 0 :
        print('Número da Conta\t\tNome do titular\t\tSaldo')
        for i in range(len(vetorConta)) :
            print(f'{vetorConta[i].numeroConta}\t\t\t{vetorConta[i].nomeTitular}\t\t\t{vetorConta[i].saldo}')
    else :
        print('Não há contas bancárias cadastradas.')

def consultar(vetorConta) :
    if len(vetorConta) > 0 :
        encontrado = False
        nomePesquisa = input('Qual o nome do titular a ser buscado? ')
        for i in range(len(vetorConta)) :
            if nomePesquisa == vetorConta[i].nomeTitular :
                encontrado = True
                print('Titular encontrado.')
                print(f'Número da conta: {vetorConta[i].numeroConta}; Nome do Titular: {vetorConta[i].nomeTitular}; Saldo: R${vetorConta[i].saldo}')
                break
        if not encontrado :
            print('Titular não encontrado')
    else :
        print('Não há contas bancárias cadastradas.')

def alterar(vetorConta) :
    if len(vetorConta) > 0 :
        numeroPesquisar = int(input('Informe o código de conta para ser alterado: '))
        for i in range(len(vetorConta)) :
            if numeroPesquisar == vetorConta[i].numeroConta :
                alteracao = int(input('''O que você deseja alterar?
                1 - Saldo
                2 - Nome do titular
                3 - Saldo e nome do titular'''))

                if alteracao == 1 :
                    vetorConta[i].saldo = float(input('Qual o novo saldo? '))
                elif alteracao == 2 :
                    vetorConta[i].nomeTitular = input('Qual o novo nome do titular? ')
                elif alteracao == 3 :
                    vetorConta[i].saldo = float(input('Qual o novo saldo? '))
                    vetorConta[i].nomeTitular = input('Qual o novo nome do titular? ')
    else :
        print('Não há contas bancárias cadastradas.')

def excluir(vetorConta) :
    if len(vetorConta) > 0 :
        for i in range(len(vetorConta)) :
            if i == 0 :
                menorSaldo = vetorConta[i].saldo
                indiceMenor = i
            if menorSaldo > vetorConta[i].saldo :
                menorSaldo = vetorConta[i].saldo
                indiceMenor = i
        vetorConta.pop(indiceMenor)
        print('Conta excluida com sucesso')
    else :
        print('Não há contas bancárias cadastradas.')
    return vetorConta

def main() :
    vetorConta = []

    print('Menu do sistema')
    print('1 - Cadastro de contas')
    print('2 - Visualização das contas')
    print('3 - Consulta à conta por nome')
    print('4 - Alteração de nome e/ou saldo de um número de conta')
    print('5 - Exclusão da conta com menor saldo')
    print('6 - Sair')

    escolha = int(input('Qual opção você deseja? '))

    while escolha < 7 :
        if escolha == 1 :
            cadastrar(vetorConta)
        elif escolha == 2 :
            visualizar(vetorConta)
        elif escolha == 3 :
            consultar(vetorConta)
        elif escolha == 4 :
            alterar(vetorConta)
        elif escolha == 5 :
            excluir(vetorConta)
        elif escolha == 6 :
            print('Fim da execução do sistema.')
            break

        print('Menu do sistema')
        print('1 - Cadastro de contas')
        print('2 - Visualização das contas')
        print('3 - Consulta à conta por nome')
        print('4 - Alteração de nome e/ou saldo de um número de conta')
        print('5 - Exclusão da conta com menor saldo')
        print('6 - Sair')

        escolha = int(input('Qual opção você deseja? '))

main()
```

A seguir são apresentados os detalhes de implementação de cada opção do menu:

1. Cadastrar clientes — não pode existir mais que um cliente com o mesmo código.

2. Relatório de clientes - listar todos os clientes cadastrados.

3. Cadastrar documentos — ao cadastrar um documento, **se o dia de pagamento for maior que o dia de vencimento, calcular o campo ‘juros’ do registro documentos (5% sobre o valor original do documento)**.

4. Relatório de documentos - listar todos os documentos cadastrados.

5. Excluir clientes — um cliente só poderá ser excluído se não existir nenhum documento associado a ele.

6. Excluir documentos individuais — por meio de seu número. Caso o documento não exista, o programa deverá mostrar a mensagem "Documento não encontrado".

7. Excluir documentos por cliente — o programa deverá informar o código do cliente e excluir todos os seus documentos. Caso o cliente não exista, deverá mostrar a mensagem "Cliente não encontrado".

8. Excluir documentos por período — o programa deverá informar o dia inicial e o dia final e excluir todos os documentos que possuam data de vencimento nesse período.

9. Alterar as informações sobre os clientes — **só NÃO altere o código do cliente**.

10. Mostrar o total de documentos de determinado cliente.
