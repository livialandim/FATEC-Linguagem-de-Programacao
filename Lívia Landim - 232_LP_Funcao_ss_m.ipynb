    r = (100 * preco_novo - 100 * preco_antigo) / preco_antigo

```
def acrescimo() :
    antigo = float(input('Informe o preço antigo do produto: '))
    atual  = float(input('Informe o preço atual do produto: '))

    percentual = (100 * atual - 100 * antigo) / antigo

    print(f'O percentual de acréscimo é igual a {percentual:.2f}%')

def main() :
    acrescimo()

main()
```

### 7. (Função sem retorno sem parâmetro) Faça uma função/método que leia três notas de um aluno e uma letra, se a letra for igual a *A*, deverá calcular a média aritimética das notas dos alunos, se for *P*, deverá calcular a média ponderada, com pesos 5, 3 e 2. A média deve ser mostrada ao final.
N1, N2 e N3 são notas.

P1, P2 e P3 são pesos.

Média ponderada = (N1 * P1 + N2 * P2 + N3 * P3 ) / (P1 + P2 + P3)

```
def media() :
    nota1 = float(input('Informe a primeira nota: '))
    nota2 = float(input('Informe a segunda nota: '))
    nota3 = float(input('Informe a terceira nota: '))

    condicao = input('Qual tipo de média será calculada? Digite A para média aritmética ou P para média ponderada ')

    if condicao == 'A' or condicao == 'a' :
        mediaA = (nota1 + nota2 + nota3) / 3

        print(f'A média aritmética das notas do aluno é {mediaA}')

    elif condicao == 'P' or condicao == 'p' :
        mediaP = ((nota1 * 5) + (nota2 * 3) + (nota3 * 2)) / (5 + 3 + 2)

        print(f'A média ponderada das notas do aluno é {mediaP}')

    else:
        print('Informe uma opção válida')

def main() :
    media()

main()
```

### 10. (Função sem retorno sem parâmetro) Faça uma função/método que leia um valor inteiro e positivo N e mostre o valor de S, obtido pelo seguinte cálculo:
###S = 1  +  1/1!  +  1/2!  +  1/3!  + ... +  1/N!

Observvação: Não pode usar vetor e função pronta da linguagem Python.

```
# Digite seu código aqui.
# S = 1
# S = 1 + 1 + 1/2! + 1/3 + 1/4!
# S = 1 + 1 + 0.5 + 0.16 + 0.041
# S = 2.66 + 0.041
# 3! = 3 * 2 * 1 = 6
# 4! = 1 * 2 * 3 * 4 = 24    1/4!   1/24  = 0.041

# N = int(input('Informe o valor de N termos: '))
def fatorial() :
    n = int(input('Informe o valor de N termos: '))
    S = 1
    termo = 1
    while n > 1 :
        fatorial = 1
        while termo <= n :
            fatorial *= termo
            termo += 1
            S = S + 1 / fatorial
        n = n - 1
    print(f'O valor de S é {S:.4f}')
def main() :
    fatorial()

main()
```
