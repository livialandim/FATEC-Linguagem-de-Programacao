### 3. (Função com retorno com parâmetro) Faça um programa contendo uma função/método que receba um valor de porcentagem de aumento via parâmetro, aplique este aumento a um salário de um funcionário, retorne e apresente o novo salário.

```
def aumento(salario, porcentagem) :
    salarioNovo = salario + (salario * (porcentagem/100))

    return salarioNovo

def main() :
    salarioAtual = float(input('Informe o salário atual de um funcionário: '))
    porcentual = float(input('Informe o aumento percentual do salário: '))

    print(f'O novo salário será R${aumento(salarioAtual, porcentual)}')

main()
```

### 4. (Função com retorno com parâmetro) Faça um programa contendo uma função/método que receba duas notas (P1, P2) calcule a média, chame outra função na main que avalie se este aluno esta aprovado ou reprovado retornando a str/string 'Aprovado' ou 'Reprovado'.

```
def avaliacao(p1, p2) :
    media = (p1 + p2) / 2

    if media >= 6 :
        print('Aprovado')
        return
    else:
        print('Reprovado')
        return

def main() :
    prova1 = float(input('Informe a nota da primeira prova: '))
    prova2 = float(input('Informe a nota da segunda prova: '))

    avaliacao(prova1, prova2)

main()
```

### 5. (Função com retorno com parâmetro) Faça um programa contendo uma função/método que receba um valor de porcentagem de aumento e um salário via parâmetro, aplique este aumento ao salário do funcionário. Na parte principal (main) do programa, na chamada da função, utilize um laço de repetição para ler 10 salários, chame a função para aplicar o aumento e gerar o novo salário, ainda dentro da estrutura de repetição acumule os novos salários e ao final apresente quanto será gasto no novo salário. Também apresente qual será a diferença entre o que se pagava para todos os funcionário e o que será pago após o aumento.

```
def aumento(salario, porcentagem) :
    salarioNovo = salario + (salario * (porcentagem/100))

    return salarioNovo

def main() :
    somaSalariosNovos = 0
    somaSalariosAtuais = 0
    diferencaSalarios = 0
    for i in range(3) :
        salarioAtual = float(input('Informe o salário atual de um funcionário: '))
        porcentual = float(input('Informe o aumento percentual do salário: '))

        salarioNovo = aumento(salarioAtual, porcentual)

        somaSalariosNovos += salarioNovo
        somaSalariosAtuais += salarioAtual

    diferencaSalarios = somaSalariosNovos - somaSalariosAtuais

    print(f'Os gastos com os novos salários será de R${somaSalariosNovos}')
    print(f'Será pago R${diferencaSalarios} a mais após o aumento')

main()
```

### 6. (Função com retorno com parâmetro) Faça um programa contendo algumas funções:
a) Crie uma função para apresentar o menu e retornar a opção/número/cálculo escolhido.

    Menu de cálculos
    1.   Número ao quadrado
    2.   Número ao cubo
    3.   Raiz do número
    4.   Raiz cúbica do número
    Qual é a opção desejada?

b) Desenvolva uma função para cada opção de cálculo.

c) A função main() chamará todas as outras.


Observação: Na chamada das funções deve-se utilizar uma estrutura de repetição, que permita diversos cálculos, quando o usuário quiser sair da calculadora digitará qualquer valor diferente dos **números** do menu.


```
def quadrado(numero) :
    elevadoQuadrado = numero ** 2

    return elevadoQuadrado

def cubo(numero) :
    elevadoCubo = numero ** 3

    return elevadoCubo

def raiz(numero) :
    raizQuadrada = numero ** (1/2)

    return raizQuadrada

def cubica(numero) :
    raizCubica = numero ** (1/3)
    return raizCubica

def opcoes() :
    print('Menu de cálculos')

    print('1. Número ao quadrado')
    print('2. Número ao cubo')
    print('3. Raiz do número')
    print('4. Raiz cúbica do número')
    print('5. Sair')

    opcao = int(input('Informe qual a operação a ser realizada: '))
    return opcao
def main() :
    decisao = opcoes()

    while decisao != 5 :
        numero = int(input('Informe o número: '))
        if decisao == 1 :
            print(quadrado(numero))
        elif decisao == 2 :
            print(cubo(numero))
        elif decisao == 3 :
            print(raiz(numero))
        elif decisao == 4 :
            print(cubica(numero))

        decisao = int(input('Informe qual a operação a ser realizada: '))

main()
```

### 7. (Função com retorno com parâmetro) Faça um programa contendo algumas funções:
a) Crie uma função para apresentar o menu e retornar **caracter** do cálculo escolhido.

    *** Minha Calculadora ***

    Digite um número.....:
    Digite outro número..:
        + Soma dois números
        - Subtrai dois números
        * Multiplica dois números
        / Divide dois números
    Qual opção deseja?

b) Desenvolva uma função para cada opção de cálculo, **mas estas não terão retorno**.

c) A função main() chamará todas as outras, passando a elas os valores digitados dos dois números.

Observação: Na chamada da função deve-se utilizar uma estrutura de repetição que permita diversos cálculos, quando o usuário quiser sair da calculadora digitará qualquer valor diferente dos **caracteres** do menu.



```
def menu() :
    print('Calculadora')
    print('+ | Soma de dois números')
    print('- | Subtração de dois números')
    print('* | Multiplicação de dois números')
    print('/ | Divisão de dois números')
    print('. | Sair')

    opcao = input('Informe a operação desejada: ')

    return opcao

def soma(n1, n2) :
    somaNumeros = n1 + n2

    print(f'A soma entre {n1} e {n2} é {somaNumeros}')
def subtracao(n1, n2) :
    if n1 > n2 :
        subtracaoNumeros = n1 - n2
    else:
        subtracaoNumeros = n2 - n1

    print(f'A subtração entre {n1} e {n2} é {subtracaoNumeros}')

def multiplicacao(n1, n2) :
    multiplicacaoNumeros = n1 * n2
    print(f'A multiplicação entre {n1} e {n2} é {multiplicacaoNumeros}')

def divisao(n1, n2) :
    if n2 != 0 :
        divisaoNumeros = n1 / n2

        print(f'A divisão entre {n1} e {n2} é {divisaoNumeros:.2f}')
    else :
        print('O segundo número precisa ser diferente de zero!')

def main() :
    opcao = menu()

    while opcao != '.' :
        n1 = int(input('Informe um número: '))
        n2 = int(input('Informe outro número: '))
        if opcao == '+' :
            soma(n1, n2)
        elif opcao == '-' :
            subtracao(n1, n2)
        elif opcao == '*' :
            multiplicacao(n1, n2)
        elif opcao == '/' :
            divisao(n1, n2)
        opcao = input('Informe a operação desejada: ')
main()
```

```
def menu() :
    print('Calculadora')
    print('+ | Soma de dois números')
    print('- | Subtração de dois números')
    print('* | Multiplicação de dois números')
    print('/ | Divisão de dois números')
    print('. | Sair')

    opcao = input('Informe a operação desejada: ')

    return opcao

def soma(n1, n2) :
    somaNumeros = n1 + n2
    return somaNumeros
def subtracao(n1, n2) :
    if n1 > n2 :
        subtracaoNumeros = n1 - n2
    else:
        subtracaoNumeros = n2 - n1

    return subtracaoNumeros

def multiplicacao(n1, n2) :
    multiplicacaoNumeros = n1 * n2
    return multiplicacaoNumeros

def divisao(n1, n2) :
    if n2 != 0 :
        divisaoNumeros = n1 / n2
        return divisaoNumeros
    else :
        print('O segundo número precisa ser diferente de zero!')
def main() :
    opcao = menu()

    while opcao != '.' :
        n1 = int(input('Informe um número: '))
        n2 = int(input('Informe outro número: '))
        if opcao == '+' :
            print('A soma entre esses números é ', soma(n1,n2))
        elif opcao == '-' :
            print('A subtração entre os números é ', subtracao(n1,n2))
        elif opcao == '*' :
            print('A multiplicação entre esses números é ', multiplicacao(n1, n2))
        elif opcao == '/' :
            print('A divisão entre esses números é igual a ', divisao(n1, n2))
        opcao = input('Informe a operação desejada: ')
main()
```

###9. (Função com retorno com parâmetro) Crie uma função para calcular/apresentar a distância entre dois pontos dados pelas coordenadas (x1, y1) e(x2, y2).distancia.png

```
def distancia(x1,y1,x2,y2) :
    calculo = (((x2 - x1)**2) + (y2 - y1)**2)**(1/2)

    return calculo

def main() :
    x1 = int(input('Informe a coordenada X do ponto 1 (x1) : '))
    y1 = int(input('Informe a coordenada Y do ponto 1 (y1) : '))

    x2 = int(input('Informe a coordenada X do ponto 2 (x2) : '))
    y2 = int(input('Informe a coordenada Y do ponto 2 (y2) : '))

    print('A distância entre os dois pontos é ', distancia(x1,y1,x2,y2))

main()
```

### 10. (Função com retorno com parâmetro) Crie uma função que verifique se um número é divisível por outro. Use o caracter %.

```
def verificacao(n1, n2) :
    return 'é divisível' if n1 % n2 == 0 else 'não é divisível'

def main() :
    n1 = int(input('Informe o número a ser verificado: '))
    n2 = int(input('Informe o outro número para verificação: '))

    print(f'O número {n1} {verificacao(n1,n2)} por {n2}')

main()
```

### 11. (Função com retorno com parâmetro) Na função main() o usuário deverá informar a quantidade de linhas e colunas, crie/chame uma função para criar e digitar os elementos de uma matriz, retorne-a. Crie/chame outra função que a calculará a soma de todas as linhas e armazenará num vetor do tamanho da quantidade de linhas, retorne-o e apresente na main().

```
# Digite aqui seu código. Exemplo de matriz 2x3
matriz = []
linha = []
def criacao(linhas, colunas) :
    for li in range(linhas) :
        linha = []
        for col in range(colunas) :
            linha.append(int(input(f'Informe o elemento da posição [{li}][{col}]')))
        matriz.append(linha)

def soma() :
    somaElementos = []

    for linha in matriz :
        somaLinhas = 0
        for i in linha :
            somaLinhas += i
        somaElementos.append(somaLinhas)
    return somaElementos

def main() :
    linhas = int(input('Informe a quantidade de linhas: '))
    colunas = int(input('Informe a quantidade de colunas: '))

    criacao(linhas, colunas)

    print(f'A soma dos elementos da linhas são, respectivamente: {soma()}')

main()
```
