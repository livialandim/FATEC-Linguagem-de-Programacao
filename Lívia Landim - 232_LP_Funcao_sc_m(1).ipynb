### 1. (Função sem retorno com parâmetro) Faça um programa contendo uma função/método que leia um número e o multiplique por 2, apresente o resultado.

```
# PASSAGEM DE PARÂMETRO POR VALOR:
# o valor da variável n que esta na def main()
# é copiado para a variável numero. Nenhuma das
# duas funções veem o conteúdo da variável de
# cada uma, ou seja, a função main() não conhece
# a variável numero, tampouco, vice-versa.

def multiplicar(numero): # número é um parâmetro
    print('Resultado', numero * 2)

def main():
    x = int(input('Digite o valor: '))
    # n é chamado de argumento na chamada de uma função
    multiplicar(x)

main()
```

### 2. (Função sem retorno com parâmetro) Faça um programa que a partir da função main(), o usuário escolha quais operações aritméticas (***) quer realizar, para isso coloque as chamadas das funções dentro de uma estrutura de repetição. Não use NENHUMA função pronta da linguagem Python.
Exemplo ***:

..............................................

Programa para cálculos

..............................................

1 - Soma

2 - Subtração

3 - Multiplicação

4 - Divisão

5 - Potência

6 - Raiz

```
def soma(n1, n2) :
    soma = n1 + n2
    print(soma)
def subtracao(n1, n2) :
    subtracao = n1 - n2
    print(subtracao)
def multiplicacao(n1, n2) :
    multiplicacao = n1 * n2
    print(multiplicacao)
def divisao(n1, n2) :
    divisao = n1 / n2
    print(divisao)
def potencia(n1, n2) :
    potencia = n1 ** n2
    print(potencia)
def raiz(n1, n2) :
    raiz = n1 ** (1/n2)
    print(raiz)
def main() :
    print('''Operações disponíveis:
    \n1- Adição
    \n2- Subtração
    \n3- Multiplicação
    \n4- Divisão
    \n5- Potenciação
    \n6- Radiciação
    \n7- Encerrar''')
    print('')
    operacao = int(input('Informe qual operação você deseja realizar: '))
    while operacao != 7 :
        numero1 = float(input('Informe um número: '))
        numero2 = float(input('Informe outro número: '))
        if operacao == 1 :
            soma(numero1, numero2)
        elif operacao == 2 :
            subtracao(numero1, numero2)
        elif operacao == 3 :
            multiplicacao(numero1, numero2)
        elif operacao == 4 :
            divisao(numero1, numero2)
        elif operacao == 5 :
            potencia(numero1, numero2)
        elif operacao == 6 :
            raiz(numero1, numero2)
        else :
            print('Informe uma opção válida')

        operacao = int(input('Informe qual operação você deseja realizar: '))

main()
```

### 4. (Função sem retorno com parâmetro) Faça uma função/método para a partir de um preço de um produto informado, calcular e apresentar o novo preço reajustado em alguma porcentagem que deve ser informada pelo usuário e será um parâmetro.

```
def calculo(preco, porcentagem) :
    produto = preco + ((preco * porcentagem) / 100)
    print(f'O novo preço do produto é: {produto:.2f}')

def main() :
    antigoPreco = float(input('Informe o preço atual do produto: '))
    reajuste = float(input('Informe a porcentagem de reajuste do preço: '))

    calculo(antigoPreco, reajuste)

main()
```

### 5. (Função sem retorno com parâmetro) Faça uma função/método para a partir de um valor inicial e um valor final realizar o acúmulo/somatório desse valores e apresentar o resultado. Não use vetor.
    Exemplo:
        a = 2

        // 2 + 3 + 4 + 5 + 6 + 7 + 8 = 35

```
def somatoria(a, b) :
    soma = 0
    for i in range(a, b + 1) :
        soma += i

    print('A somatória entre esses valores é ', soma)
def main() :
    a = int(input('Informe o valor inicial dessa somatória: '))
    b = int(input('Informe o valor final dessa somatória: '))

    somatoria(a, b)

main()
```

### 6. (Função sem retorno com parâmetro) Faça uma função/método para calcular a tabuada de um número informado pelo usuário.

```
def tabuada(numero) :
    print('Tabuada do ', numero)
    valor = 0
    while valor < 11 :
        multiplicacao = numero * valor
        valor += 1

        print(multiplicacao)

def main() :
    num = int(input('Informe um número para calcular sua tabuada : '))

    tabuada(num)
main()
```

###7. (Função sem retorno com parâmetro) Faça uma função/método para calcular a tabuada de um ***intervalo*** informado pelo usuário, por exemplo, tabuada do 3 ao 8.

```
def tabuada(numero, numero2) :
    while numero <= numero2:
        valor = 0
        print('Tabuada do ', numero)
        while valor < 11 :
            multiplicacao = numero * valor
            valor += 1

            print(multiplicacao)
        numero += 1

def main() :
    num = int(input('Informe um número para abrir o intervalo de tabuadas : '))
    num2 = int(input('Informe outro número para fechar o intervalo de tabuadas: '))

    tabuada(num, num2)
main()
```

### 11. (Função sem retorno com parâmetro) Faça uma função/método para verificar e contar quantas letras **a** tem numa frase. Não use NENHUMA função pronta da linguagem Python.

```
def verificacao(frase) :
    contador = 0
    for a in frase:
        if a == 'a' or a == 'A' :
            contador +=1

    print('A quantidade de letras A presentes na frase é ', contador)
def main() :
    frase = input('Informe uma frase: ')

    verificacao(frase)

main()
```

Programa para Gerenciamento de Estoque de Produtos

1 - Cadastro do nome, quantidade e preço do produto

4 - Aplicação de reajuste de aumento 5%

5 - Alteração de nome de produto de camisa para blusa

**Explicação:**

**Opção 1:** o preenchimento/digitação dos dados, deve ocorrer um vetor para nome, outro para quantidade e mais um para preço

**Opção 2:** siga o exemplo de apresentação dos produtos (***). Utilize os recursos de formatação de str passados no slide de comandos para ter a visualização mostrada

**Opção 3:** apresente apenas os produtos que estão <= a 3 (quantidade mínima)

**Opção 4:** aplique reajuste de aumento 5% aos produtos que iniciam startswith(), função Python, coma letra c

**Opção 5:** a alteração, replace(), de nome de produto de camisa para blusa, so deve ocorrer nesta palavra, ou seja, branca e amarela continuam.

**Opção 6:** use a função da str, capitalize() para colocar todos os nomes de produtos com a primeira letra maiúscula


```
(***)
Nome                    Quantidade            Preço R$

Camisa branca               102                  59.80
Vestido longo azul           10                 310.79
Colar prata                   2                1200.91
Camisa amarela               30                  61.99
...
```

```
elementos = []
quantidade = []
preco = []
def cadastro() :
    for i in range(3) :
        elementos.append(input('Informe o nome do produto: '))
        quantidade.append(int(input('Informe a quantidade do produto: ')))
        preco.append(float(input('Informe o preço do produto')))

def apresentacao() :
    for i in range(len(elementos)) :
        print(elementos[i],' | ', quantidade[i],' | ', preco[i])

def minima() :
    for i in range(len(elementos)) :
        if quantidade[i] <= 3 :
            print('----- Produtos com quantidade mínima -----')
            print(elementos[i],' | ', quantidade[i],' | ', preco[i])

def aumento() :
    for i in range(len(elementos)) :
        if elementos[i].startswith('c') or elementos[i].startswith('C') :
            preco[i] = preco[i] + (preco[i] * 0.05)


def substituicao() :
    for i in range(len(elementos)) :
        elementos[i] = elementos[i].replace('Camisa','Blusa')


def maiuscula() :
    for i in range(len(elementos)) :
        elementos[i] = elementos[i].capitalize()

def main() :
    print('Programa para Gerenciamento de Estoque de Produtos')
    print('-' * 50)
    print('1 - Cadastro do nome, quantidade e preço do produto')
    print('2 - Apresentação de todos os produtos')
    print('3 - Apresentação de estoque mínimo')
    print('4 - Aplicação de reajuste de aumento 5%')
    print('5 - Alteração de nome de produto de camisa para blusa')
    print('6 - Capitalize todos os nomes de produtos')
    print('7 - Sair')

    opcao = int(input('Informe a opção desejada: '))

    while opcao != 7 :
        if opcao == 1 :
            cadastro()
        elif opcao == 2 :
            apresentacao()
        elif opcao == 3 :
            minima()
        elif opcao == 4 :
            aumento()
        elif opcao == 5 :
            substituicao()
        elif opcao == 6 :
            maiuscula()

        opcao = int(input('Informe a opção desejada: '))

main()
```

Crie outra função que fará a inserção de mais um filho para este funcionário no vetor, apresente após a verificação a relação de nomes de funcionários e a quantidade de filhos de todos.

```
# Digite seu código aqui.
# os nomes dos funcionários devem ser digitados em maiúsculo,
# para isso use a função da str upper()
#...........................................................

funcionario = []
filhos = []

def verificacao(funcionario, filhos) :
    for i in range(3) :
        funcionario.append(input('Informe o nome do funcionário: '))
        filhos.append(int(input('Informe a quantidade de filhos do funcionário: ')))
        funcionario[i] = funcionario[i].upper()

def adicionar(funcionario, filhos) :
    nascimento = input('Ocorreu o nascimento de algum filho? (S/N)')


    while nascimento == 's' or nascimento == 'S' :
        qual = input('Qual funcionário ganhou filho? ')
        qual = qual.upper()
        for i in range(len(funcionario)) :
            if qual == funcionario[i] :
                filhos[i] += 1
        nascimento = input('Ocorreu o nascimento de algum filho? (S/N)')

def apresentacao(funcionario, filhos) :
    for i in range(len(funcionario)) :
        print('Funcionário: ', funcionario[i], '| Quantidade de filhos: ', filhos[i])

def main() :

    verificacao(funcionario, filhos)

    adicionar(funcionario, filhos)

    apresentacao(funcionario,filhos)
main()
```

###15. De acordo com a figura a seguir crie uma função que valide a senha, seguindo as regras demonstradas. Digite a senha na função main(), chame função de validação e apresente se a senha está correta ou incorreta.
senha.png

```
def validacao(senha) :
    numero = False
    minusculo = False
    maiusculo = False
    especial = False
    if len(senha) >= 8 :
        for caractere in senha :
            if caractere.isdigit() :
                numero = True

            if caractere.islower() :
                minusculo = True

            if caractere.isupper() :
                maiusculo = True

            if not caractere.isalnum() :
                especial = True

    if numero == True and minusculo == True and maiusculo == True and especial == True :
        print('A senha está correta')
    else :
        print('A senha está incorreta')

def main() :
    senha = input('Informe a senha: ')

    validacao(senha)

main()
```
